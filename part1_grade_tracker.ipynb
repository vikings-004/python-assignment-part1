## Task 1 — Data Cleaning

```python
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

cleaned_students = []

for s in raw_students:
    name = s["name"].strip().title()
    roll = int(s["roll"])
    marks = [int(x) for x in s["marks_str"].split(", ")]

    cleaned_students.append({"name": name, "roll": roll, "marks": marks})

    words = name.split()
    valid = all(w.isalpha() for w in words)

    if valid:
        print("✓ Valid name -", name)
    else:
        print("✗ Invalid name -", name)

    print("================================")
    print("Student :", name)
    print("Roll No :", roll)
    print("Marks   :", marks)
    print("================================")
    print()

for s in cleaned_students:
    if s["roll"] == 103:
        print(s["name"].upper())
        print(s["name"].lower())
```

## Task 2 — Marks Analysis

```python
student_name = "Ayesha Sharma"
subjects = ["Math", "Physics", "CS", "English", "Chemistry"]
marks = [88, 72, 95, 60, 78]

for i in range(len(subjects)):
    m = marks[i]
    if m >= 90:
        grade = "A+"
    elif m >= 80:
        grade = "A"
    elif m >= 70:
        grade = "B"
    elif m >= 60:
        grade = "C"
    else:
        grade = "F"
    print(subjects[i], ":", m, ":", grade)

print()
total = sum(marks)
average = round(sum(marks) / len(marks), 2)
print("Total :", total)
print("Average :", average)
print("Highest :", subjects[marks.index(max(marks))], max(marks))
print("Lowest :", subjects[marks.index(min(marks))], min(marks))
```

```python
# while loop to add new subjects
new_count = 0

while True:
    sub = input("Enter subject name (or done to stop): ")
    if sub == "done":
        break
    m = input("Enter marks (0-100): ")
    if not m.isnumeric():
        print("Warning: enter a valid number")
        continue
    if int(m) < 0 or int(m) > 100:
        print("Warning: marks should be between 0 and 100")
        continue
    subjects.append(sub)
    marks.append(int(m))
    new_count += 1

print()
print("New subjects added :", new_count)
print("Updated average :", round(sum(marks) / len(marks), 2))
```

## Task 3 — Class Performance Summary

```python
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

averages = []
pass_count = 0
fail_count = 0

print("Name              | Average | Status")
print("----------------------------------------")

for student in class_data:
    name = student[0]
    marks = student[1]
    avg = round(sum(marks) / len(marks), 2)
    averages.append((name, avg))

    if avg >= 60:
        status = "Pass"
        pass_count = pass_count + 1
    else:
        status = "Fail"
        fail_count = fail_count + 1

    print(name, "|", avg, "|", status)

print()
print("Passed :", pass_count)
print("Failed :", fail_count)

topper_name = averages[0][0]
topper_avg = averages[0][1]

for i in range(len(averages)):
    if averages[i][1] > topper_avg:
        topper_name = averages[i][0]
        topper_avg = averages[i][1]

print("Topper :", topper_name, topper_avg)

total_avg = 0
for i in range(len(averages)):
    total_avg = total_avg + averages[i][1]

class_avg = round(total_avg / len(averages), 2)
print("Class Average :", class_avg)
```

## Task 4 — String Manipulation

```python
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

# step 1
clean_essay = essay.strip()
print("Stripped:", clean_essay)
print()

# step 2
print("Title Case:", clean_essay.title())
print()

# step 3
count = clean_essay.count("python")
print("Python count:", count)
print()

# step 4
replaced = clean_essay.replace("python", "Python 🐍")
print("Replaced:", replaced)
print()

# step 5
sentences = clean_essay.split(". ")
print("Sentences:", sentences)
print()

# step 6
for i in range(len(sentences)):
    s = sentences[i]
    if s.endswith(".") == False:
        s = s + "."
    print(str(i + 1) + ".", s)
```
